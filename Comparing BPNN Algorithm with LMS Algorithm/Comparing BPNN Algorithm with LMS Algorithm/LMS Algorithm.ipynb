```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_input = [ 
            np.array([1, 1]), 
            np.array([1, 0]),
            np.array([0, 1]),
            np.array([1, 0])
]

data_output = [
            np.array([0]), 
            np.array([1]),
            np.array([1]),
            np.array([0])  

]

w = np.array([0.5, 0.3])
bias = 0.1
alpha = 0.5
Epoch = 2
```

```python
def adeline(weight,d,b):
  return np.dot(weight, d.transpose())+b
```

```python
def calculate_error(target, y):
  return (1/2)*((target-y)**2)
```

```python
for i in range(Epoch):
  print('Epoch ' + str(i+1))
  for j in range(4):
    yu = adeline(w, data_input[j], bias)
    error = calculate_error(data_output[j], yu)
    w = w - (alpha * (yu - data_output[j]) * data_input[j])
    bias = bias + alpha * (yu - data_output[j])
    print("y(u) : " , yu)
    print("Error : " , error)
    print("New Weight : " , w)
    print("New Bias : " , bias)
    print('\n')

```
